# Regressão Linear - Teoria

A regressão linear é uma técnica de Machine Learning (Aprendizado de Máquina) - um dos campos de Inteligência Artificial - que consiste em tentar descobrir uma relação linear entre os dados utilizados no treinamento da máquina para poder prever/inferir algo sobre esses dados. 

Em outras palavras, é tentar encontrar uma relação de caráter linear entre as variáveis dadas como entrada e as variáveis dadas como saída. Veja o gŕafico abaixo que ilustra em um plano cartesiano esse objetivo:

```python
import torch
import matplotlib.pyplot as plt
```

```python
X = torch.arange(-5, 5, 0.1).view(-1, 1)
func = -5 * X
Y = func + 0.7 * torch.randn(X.size())
```

```python
plt.plot(X.numpy(), Y.numpy(), 'b+', label='Y')
plt.plot(X.numpy(), func.numpy(), 'r', label='func')
plt.xlabel('x - Dados')
plt.ylabel('y - Rótulos')
plt.legend()
plt.grid('True', color='y')
plt.show()
```
